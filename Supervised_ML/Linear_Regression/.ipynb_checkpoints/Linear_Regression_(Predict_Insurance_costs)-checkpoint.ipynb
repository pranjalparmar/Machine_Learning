```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error as mae,mean_squared_error as mse,r2_score
```

```python
data_path = "./insurance.csv"
```

```python
df = pd.read_csv(data_path)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
le = LabelEncoder()
```

```python
df['smoker'] = le.fit_transform(df['smoker'])
```

```python
sns.pairplot(df, kind = 'scatter')
```

```python
sns.lmplot(data=df, x = 'age', y = 'charges')
```

```python
sns.scatterplot(data=df, x = 'age', y = 'charges')
```

```python
sns.boxplot(data=df, x = 'smoker', y = 'charges')
```

```python
X = df[['age','bmi','smoker']]
y = df['charges']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
cdf = pd.DataFrame(lr.coef_, X.columns, columns=['Coefficient'])
print(cdf)
```

```python
predictions = lr.predict(X_test)
```

```python
predictions
```

```python
sns.scatterplot(x=predictions, y=y_test)
```

```python
print("Mean Absolute Error:",mae(y_test,predictions))
print("Mean Squared Error:",mse(y_test,predictions))
print("Root Mean Squared Error:",np.sqrt(mse(y_test,predictions)))
print("R2_score:",r2_score(y_test,predictions))
```

```python
residuals = y_test - predictions
```

```python
residuals
```

```python
sns.displot(residuals, bins = 25, kde = True)
```

```python
import pylab
import scipy.stats as stats

stats.probplot(residuals, dist = "norm", plot = pylab)
pylab.show()
```
