```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from skimpy import skim
```

```python
df = sns.load_dataset('tips')
```

```python
df.info()
```

```python
skim(df)
```

```python
df.columns
```

```python
X = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y = df['total_bill']
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train
```

### Feature Encoding (Label Encoding & One Hot Encoding)

```python
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] = le2.fit_transform(X_train['smoker'])
X_train['time'] = le3.fit_transform(X_train['time'])
```

```python
X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] = le2.transform(X_test['smoker'])
X_test['time'] = le3.transform(X_test['time'])
```

```python
print("Categories in train:", X_train['sex'].unique())
print("Categories in test:", X_test['sex'].unique())
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])], remainder='passthrough')
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
X_train = ct.fit_transform(X_train)
```

```python
X_test = ct.transform(X_test)
```

```python
X_test
```

### Support Vector Regressor

```python
from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import classification_report, r2_score, mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

### Hyperparameter Tuning with Grid Search CV

```python
from sklearn.model_selection import GridSearchCV

params = {
    'C' : [0.1,1,10,100,1000],
    'gamma' : [1,0.1,0.01,0.001,0.0001],
    'kernel' : ['linear','sigmoid','rbf']
}
```

```python
grid = GridSearchCV(estimator=SVR(),param_grid=params,refit=True, verbose=3)

grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred_grid = grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y_test,y_pred_grid))
print(mean_absolute_error(y_test,y_pred_grid))
```
